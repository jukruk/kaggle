```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
train.head()
```

```python
train.drop(columns=['id'], inplace=True)
```

```python
train.info()
```

```python
object_columns = list(train.dtypes[train.dtypes=='object'].index)
df_vc = pd.DataFrame()
for column in object_columns:
    df_vc = pd.concat([df_vc, train[column].value_counts().reset_index()], axis=1)
df_vc
```

```python
train.loc[(train.Gender=='Male'), 'Gender'] = 1
train.loc[(train.Gender=='Female'), 'Gender'] = 0
train.loc[(train.family_history_with_overweight=='yes'), 'family_history_with_overweight'] = 1
train.loc[(train.family_history_with_overweight=='no'), 'family_history_with_overweight'] = 0
train.loc[(train.FAVC=='yes'), 'FAVC'] = 1
train.loc[(train.FAVC=='no'), 'FAVC'] = 0
train.loc[(train.SMOKE=='yes'), 'SMOKE'] = 1
train.loc[(train.SMOKE=='no'), 'SMOKE'] = 0
train.loc[(train.SCC=='yes'), 'SCC'] = 1
train.loc[(train.SCC=='no'), 'SCC'] = 0
train.loc[(train.NObeyesdad=='Insufficient_Weight'), 'NObeyesdad'] = 0
train.loc[(train.NObeyesdad=='Normal_Weight'), 'NObeyesdad'] = 1
train.loc[(train.NObeyesdad=='Overweight_Level_I'), 'NObeyesdad'] = 2
train.loc[(train.NObeyesdad=='Overweight_Level_II'), 'NObeyesdad'] = 3
train.loc[(train.NObeyesdad=='Obesity_Type_I'), 'NObeyesdad'] = 4
train.loc[(train.NObeyesdad=='Obesity_Type_II'), 'NObeyesdad'] = 5
train.loc[(train.NObeyesdad=='Obesity_Type_III'), 'NObeyesdad'] = 6

train['Gender'] = train['Gender'].astype('Int64')
train['family_history_with_overweight'] = train['family_history_with_overweight'].astype('Int64')
train['FAVC'] = train['FAVC'].astype('Int64')
train['SMOKE'] = train['SMOKE'].astype('Int64')
train['SCC'] = train['SCC'].astype('Int64')
train['NObeyesdad'] = train['NObeyesdad'].astype('Int64')

train.dtypes
```

```python
train.describe()
```

```python
object_columns = list(train.dtypes[train.dtypes=='object'].index)
df_vc = pd.DataFrame()
for column in object_columns:
    df_vc = pd.concat([df_vc, train[column].value_counts().reset_index()], axis=1)
df_vc
```

```python
def onehot(df:pd.DataFrame, column:str) -> pd.DataFrame:
    one_hot_df = pd.DataFrame()
    df_u = df[column].unique()
    for un in df_u:
        one_hot_df[f'{column}_{un}'] = np.where(df[column]==un,1,0)
    return one_hot_df
```

```python
one_hot_df = pd.DataFrame()
for column in object_columns:
    one_hot_df = pd.concat([one_hot_df, onehot(train, column)], axis=1)
one_hot_df.head()
```

```python
df_fin = train.drop(object_columns, axis=1)
df_fin = pd.concat([df_fin, one_hot_df], axis=1)
df_fin
```

# First model

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
y = df_fin.NObeyesdad
X = df_fin.drop(['NObeyesdad'],
             axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)
```

```python
model1 = XGBClassifier(objective='multi:softmax', 
                        num_class=7,)
model1.fit(X_train, y_train)
print(f"Accuracy score: {accuracy_score(y_test, model1.predict(X_test))}")
```

# Grid search

```python
model = XGBClassifier(objective='multi:softmax', num_class=7,)

param_grid = {
    'max_depth': [3, 4, 5],
    'learning_rate': [0.1, 0.01, 0.05],
    'n_estimators': [50, 100, 200],
    'subsample': [0.8, 0.9, 1.0],
    'colsample_bytree': [0.8, 0.9, 1.0],
}

grid_search = GridSearchCV(estimator=model, param_grid=param_grid, scoring='accuracy', cv=3, verbose=1)

grid_search.fit(X_train, y_train)
print("Najlepsze parametry:", grid_search.best_params_)
y_pred = grid_search.predict(X_test)

# Ocena dokÅ‚adnoÅ›ci modelu
accuracy = accuracy_score(y_test, y_pred)
print(f'DokÅ‚adnoÅ›Ä‡ modelu: {accuracy}')
```
